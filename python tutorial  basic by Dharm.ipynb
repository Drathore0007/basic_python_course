### Tuple

Tuples are same as list but they are immutable.

Advantage of using tuple over list is 
* tuples are faster to access
* takes less memory than list
  
Note: Use tuples when you don't require to update your container everythime.

```python
""" Tuple """

print("Defining a tuple")
t = (2,5,2,"hello", "team", False, True, 5.0)
print(t, "\n")  

print("len() returns a lenght of a tuple.")
print(len(t), "\n") 

print("tuple[index] returns a vlaue at that index")
print(t[0])   
print(t.__getitem__(0))
print("\n")


print("tuple[starting index: ] retuns a list elements.")
print(t[0:], "\n") 

print("tuple[:] returns all elements from list")
print(t[:], "\n")

print("tuple[starting index: end index] retuns a list element where end index element is not returned.") 
print(t[0:2], "\n") 

print("tuple[:ending index] default starting index is 0 here ")
print(t[:-1], "\n")

print("tuple_object.index(element) returns a index of that element.")
print(t.index(5), "\n")  

print("tuple_object.count(element) returns number of time it has occurancy in tuple")
print(t.count("hello"), "\n") 

print("Available functions for tuple: ")
print(dir(t))
```

###  Finding a mutual friend in FB

<p> Problem: suppose there are two friends A and B. A has friends [B,C,D,E,F] and B has friends like [A,C,E,G,H]. Now mutual friends between them will be (A,B):[C,E].</p>
<p> Now if we have data like below
    <code> "Dharm":["Aman","Jatin","Anshika","Girish","Hitu","Madhu"], 
"Aman":["Mahak","Jatin","Anshika","Amit","Dharm"], 
"Anshika":["Aman","Jatin","madhu","divya","Dharm"], 
"Girish":["Jatin","mahakanta","Mahendra","Dharm"],
 "Jatin":["Dharm","Aman","Anshika","Ankit"] </code> </p>
 
 <p > Find mutual friend between each of them </p>
    
 

```python
# solution code 1

FB = {"Dharm":["Aman","Jatin","Anshika","Girish","Hitu","Madhu"], 
"Aman":["Mahak","Jatin","Anshika","Amit","Dharm"], 
"Anshika":["Aman","Jatin","madhu","divya","Dharm"], 
"Girish":["Jatin","mahakanta","Mahendra","Dharm"],
 "Jatin":["Dharm","Aman","Anshika","Ankit"] }   # creating a dictionary FB


def mapper(disc):    # mapper function defination 
    data = disc
    mutualfriends={  }    # creating an empty dictionary 
    for f1 in FB:        #taking each person from FB dictionary
        personf = FB[f1]      #stroring person friend in person list
        for f2 in FB:
            if f1 != f2:      # checking if person is not person itself.
                friendf = FB[f2]
                key = [f1, f2] # creating a tuple of (person , friend)
                value = []    #creating an empty list 
                
                value.append(personf)   
                value.append(friendf)
                key= sorted(key)
                key =tuple(key)   
                
                mutualfriends[key]= value  # creating a key and value in dictionary         

    return mutualfriends  # returning final mutual frind dictionary


def reducer(dics):
    data = dics     #access mutualfriends dictionary as data in reducer
    mutualfriend ={}   # defining a dictionary to store final result.
    
    for k in data:
        key = k
        friends = []
        value = data[k]
        personfriends=value[0] #storing first value as personfriends
        friendfriends= value[1] #storing first value as friendfriends
        
        for i in range(len(personfriends)):
            for j in range(len(friendfriends)):
                if personfriends[i]==friendfriends[j]:  # if friend in personf matches friendf
                    friends.append(personfriends[i])    #append that value in friends list

        mutualfriend[key]= friends
                   

    return mutualfriend  # return the final dictionary


fbmap = mapper(FB) # calling maper fuction and passig data set

fbreduce = reducer(fbmap) # calling reduce function and passing mapper output

print("Mapper output: ") 
print(fbmap)

print ("\n")

print("Reducer Output: ")  
print(fbreduce)  
```
